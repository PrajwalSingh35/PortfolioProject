# Exploratory Data Analysis

### Gender

```python
#plotting a bar chart for gender and its count
ax= sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting bar chart for gender vs amount
gen_sales = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount',data= gen_sales)
```

###### From above graph we conclude that most of the customers are female and also the amount spent my females is greater than male

### Age Group

```python
# plot of total customers based on age group and hue we have used to categorise it based on gender
ax= sns.countplot(x='Age Group', data=df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plot of Total Amount vs Age Group
age_sales= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.barplot(x='Age Group', y='Amount',data= age_sales)
```

##### From above graphs we conclude that most of the buyers are females from age group 26-35

### State

```python
# total orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,4)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
# used head to give top 10 states only and fig size is used to adjust the size of chart
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,4)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

##### From above charts we conclude that most of the orders are from Uttar pradesh, Maharashtra and Karnataka respectively

### Products

```python
sns.set(rc={'figure.figsize':(24,5)})
ax = sns.countplot(x='Product_Category', data=df)

for bar in ax.containers:
    ax.bar_label(bars)
```

```python
# total amount spent on product categories
category_sales=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(26,5)})
sns.barplot(x='Product_Category', y='Amount', data=category_sales)
```

##### From above graph we can see that the most sold products are from Food and Clothing categories respectively

```python
# top 10 most ordered product with respect to the Product ID
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,4)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
